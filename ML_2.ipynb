```python
import Orange

# Load built-in Iris dataset
data = Orange.data.Table("iris")

# CN2 rule learner
print("\n=== CN2 RULES ===")
cn2 = Orange.classification.rules.CN2Learner()(data)
for r in cn2.rule_list:
    print(r)

# FOIL-like rule learner
print("\n=== FOIL-LIKE RULES ===")
foil = Orange.classification.rules.CN2SDUnorderedLearner()(data)
for r in foil.rule_list:
    print(r)
```
