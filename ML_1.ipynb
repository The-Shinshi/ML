```python
dataset = [['Sunny','warm','normal','strong','warm','same','yes'],
['Sunny','warm','high','strong','warm','same','yes'],
['rainy','cold','normal','strong','warm','change','no'],
['Sunny','warm','high','strong','cool','change','yes']]
```

```python
def find_s(dataset):
    hypothesis = ['0'] * (len(dataset[0]) - 1)

    for example in dataset:
        attributes, label = example[:-1], example[-1]

        if label == "yes":
            for i in range(len(attributes)):
                if hypothesis[i] == '0':
                    hypothesis[i] = attributes[i]
                elif hypothesis[i] != attributes[i]:
                    hypothesis[i] = '?'

    return hypothesis
print("Find_s hypothesis",find_s(dataset))
```

```python
def candidate_elimination(dataset):
    n = len(dataset[0]) - 1

    S = ['0'] * n

    G = [['?'] * n]

    def consistent(h, x):
        return all(h[i] == x[i] or h[i] == '?' for i in range(len(h)))

    for example in dataset:
        x = example[:-1]
        y = example[-1]

        if y == 'yes':
            for i in range(n):
                if S[i] == '0':
                    S[i] = x[i]
                elif S[i] != x[i]:
                    S[i] = '?'

            G = [g for g in G if consistent(g, x)]

        else:
            new_G = []
            for g in G:
                for i in range(n):
                    if g[i] == '?' and S[i] != x[i]:
                        new_h = g.copy()
                        new_h[i] = S[i]
                        new_G.append(new_h)

            # Replace G with its updated list
            G = new_G

    return S, G


S, G = candidate_elimination(dataset)
print("Most Specific Hypothesis S:", S)
print("Most General Hypothesis Set G:", G)
```
